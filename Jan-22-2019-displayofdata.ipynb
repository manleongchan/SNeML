# This code loads the training/testing/validing data generated using SNe_shift_waveform.ipynb 
# The data is generated using a sampling rate = 8192 Hz - so the frequency range of the ASD used is [1Hz, 4096Hz]
# The data shown in this page include whitened waveform at SNR equal to 25, 23, 21, 19, 17, 15, 13, 11, 9, 7, 5

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate, interpolate, signal, optimize, stats
import pickle
np.set_printoptions(edgeitems=30, linewidth=160)
```

```python
#load the data
with open("showdata.pkl") as f:
    ts, show_data = pickle.load(f)
```

# The plots on the left in the following cells show the whitened training data 
# The plots on the right in the following cells show the whitened training data overlayed by the whitened waveform 
### This cell shows the training data where the optimal SNR of the waveform is = 25


```python
fig = plt.figure(figsize=(22,11), dpi= 100, facecolor='w', edgecolor='k')
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.07, hspace=None)
nrows = 1
ncols = 2
fontsize = 20

ax = fig.add_subplot(nrows, ncols, 1)
ax.plot(ts, show_data[0], linewidth = 1, color = 'b', label = 'training data (Whitened)')

ax.legend(loc ='upper right', fontsize=  13)
ax.grid()
ax.set_xlim([ts[0], ts[-1]])
ax.set_ylim([-5, 5])

ax.set_xlabel('Time(s)', fontsize = fontsize)
ax.set_ylabel('Whitened Strain', fontsize = fontsize)


for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')

ax = fig.add_subplot(nrows, ncols, 2)
ax.plot(ts, show_data[0], linewidth = 1, color = 'b', label = 'training data (Whitened)')
ax.plot(ts, show_data[1], linewidth = 1, color = 'r', label = 'Waveform without noise (Whitened)')

ax.legend(loc ='upper right', fontsize=  13)
ax.grid()
ax.set_xlim([ts[0], ts[-1]])
ax.set_ylim([-5, 5])

ax.set_xlabel('Time(s)', fontsize = fontsize)

for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
plt.show()
#fig=plt.figure(figsize=(8, 6.5), dpi= 100, facecolor='w', edgecolor='k')
```

### The following plot shows that the a simulated whiten data stream with only whiten noise and 
### without any gravitational wave signal should have the amplitudes distribute as a Gaussian distribution with mean = 0 and variance = 1

```python
fig = plt.figure(figsize=(22,11), dpi= 100, facecolor='w', edgecolor='k')
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.12, hspace=None)
nrows = 1
ncols = 2
fontsize = 20

ax = fig.add_subplot(nrows, ncols, 1)

ax.plot(ts, show_data[22], linewidth = 1, color = 'b', label = 'training data (only Gaussian noise)')

ax.legend(loc ='upper right', fontsize=  13)
ax.grid()
ax.set_xlim([ts[0], ts[-1]])
ax.set_ylim([-5, 5])

ax.set_xlabel('Time(s)', fontsize = fontsize)
ax.set_ylabel('Whitened Strain', fontsize = fontsize)


for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
    
ax = fig.add_subplot(nrows, ncols, 2)

gd = stats.norm(0,1)
x = np.linspace(-10, 10, 1000)
y = gd.pdf(x)
ax.plot(x,y, color = 'b', label = 'Gaussian distribution with unit variance and zero mean')
ax.hist(show_data[22], bins = 30, density=True, label = 'the distribution of the amplitude of a data stream with only gaussian noise', color= 'r')
ax.legend(loc ='upper right', fontsize=  13)
ax.set_ylim([0,0.5])
ax.set_ylabel('Probability density', fontsize = 20)


for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('normal')
    
    
    
plt.show()
```
