### 1. Coherent network analysis of gravitational waves from three-dimensional core-collapse supernova models by Hayama, et al 2015

## Figure 1. the locations of the waveforms in the hrss-fc in Hayama, et al 2015 and of those used to train the CNN

### 4.New method to observe gravitational waves emitted by core collapse supernovae


### These plots show the efficiency curve as a function of SNR if the threshold is set to be a constant.
### This threshold is used when CNN tries to determine what class the input belongs to.
### When training a CNN to determine the class of an input, the output is probabilities that the input belongs to each of the classes.
### For the CNN to determine that an input belongs to a class, the corresponding probability has to be larger than a threshold. 
### In this case, the threshold is 50%.
### To compare the result (the left panel) in Astone, et al, 2018,
### one should focuse on the deep red dotted line "True alarm probability for all signal classes"
### and the red dashed line  "True alarm probability for all classes"
### This is because it is not entirely clear which curve is actually meant in the paper, so two were produced.
### To compare to the red line in the same figure in that paper, one should focus on the blue dashed line "False alarm rate for signals"
### These results are shown as a function of SNR of the signals so they are not directly comparable to the results above.
### The network for both panels is HLV
### It can be seen that at SNR = 10, the result in the left panel suggests that their method can achieve True alarm probability close to 90%, while the result on the right is just close to 70%
### This is the reason that I did not consider my results were publisable, but now by comparing other results in the literature, I begin to doubt what actualy is meant by their result.
### perhaps it is because they use "phenomenological waveforms"?
