# This page shows the 
# 1. True alarm probability (TAP) vs false alarm probability(FAP) where the threshold for the CNN is 0.5 (The original Figure 7 in the paper)
# 2. TAP vs FAP where the threshold for the CNN is 0.85 (Threshold is inceased to reduce FAP to below 0.3%)
#
# The network is H L V K and H+ L+ V K

```python
import Image
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import numpy as np
```

# Figure 1. TAP VS FAP for neutrino driven mechanism
# Left panel: TAP vs FAP where the threshold for the CNN is 0.5 (The original Figure 7 in the paper)
# Right panel: TAP vs FAP where the threshold for the CNN is 0.85



```python
img = mpimg.imread('ROC_fixed_neutrino.png')

img2 = mpimg.imread('revised_fixed_threshold_neutrino.png')


fig, ax1 = plt.subplots(1,2, figsize=(41.,41), sharex = False)
gs1 = gridspec.GridSpec(4, 4)
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1], wspace=0.0, hspace=-10.0, top=0.95, bottom=0.05, left=0.17, right=0.845) 
ax1[0].imshow(img)
ax1[1].imshow(img2)

for i in range(len(ax1)):
    ax1[i].patch.set_visible(False)
    ax1[i].axis('off')
    #ax1[i].set_aspect('equal')
```
