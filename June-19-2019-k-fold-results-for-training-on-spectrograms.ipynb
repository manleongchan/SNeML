## The network was trained to classify which class among "Neutrino-driven, Magneto-rotational, Background" the input data belonged to.

## Data generation procedure
### 1. 110 magnetorational waveforms were randomly drawn, and all 22 neutrino waveforms were used.
### 2. These original waveforms were first passed through a low filter to filter out low frequency component (< 11 Hz)
### 3. A tukey window with alpha = 0.08 was then applied to every waveform. 
### 4. The waveform with the longest length was then identified, each of the rest waveforms was zero padded to this length.
### 5. 22 neutrino waveforms were then duplicated 5 times so that the number of them was equal to the number of magnetorotational waveforms.
### 6. 110 magnetorotational waves + 110 neutrino waveforms = 220 waveforms were scaled to a fixed SNR, e.g.: 4. 
### 7. 66000 noise realisations were generated, each of which was 1.7 times as long as the waveforms.
### 8. The 220 waveforms were then duplicated 200 times making them 44000 in total.
### 9. 44000 waveform data were add to 44000 of the 66000 noise realisations, with the start of the signals being randomised. 
### 10. All 66000 samples were then transformed to TF data (spectrograms).
### 11. Since different length of window and different starting position of the signal could produce a difference in the values of the TF data,
###       a normalisation was applied.
### 12. The normalisation was applied by dividing the TF data by the largest value in the TF data, so that each TF sample had their values in the range of (0, 1]
### 13. After the normalisation, the 66000 TF data samples were then used for traning.

## Training
### The training process was similar to what had been previously done.
### A 10 fold cross validation analysis was carried out for each SNR from 4 to 10. 
### When the CNN had finished training for 1 SNR, it would turn to the next SNR by repeating step 6 to 13 defined above.
### Some of the training parameters: 
### 1. epochs: 30, batch size :64, learning rate = 0.0002
