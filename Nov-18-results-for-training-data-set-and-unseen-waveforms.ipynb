```python
import Image
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import numpy as np
```

# Figure 1. Detection/Classification efficiency for the two explosion mechanisms. 
# Left panel: magnetorotational explosion mechanism. Right panel: neutrino-driven mechanism
## Solid lines are for network : H+ L+ V K; Semi-dotted lines are for: H L V K 

```python
img = mpimg.imread('mag_eff.png')
img2 = mpimg.imread('Neutrino_eff.png')

fig, ax1 = plt.subplots(1,2, figsize=(41.,41), sharex = False)
gs1 = gridspec.GridSpec(4, 4)
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1], wspace=0.0, hspace=-10.0, top=0.95, bottom=0.05, left=0.17, right=0.845) 
ax1[0].imshow(img)
ax1[1].imshow(img2)

for i in range(len(ax1)):
    ax1[i].patch.set_visible(False)
    ax1[i].axis('off')
    #ax1[i].set_aspect('equal')
```
