## This page shows the comparisons of results from other studies and from the CNN.
## The waveforms used in this comparison were not used to train/validate/test the CNN at any point.

## Procedure

### 1. The waveforms used in this page first went through the preparation procedure similar to that for the waveforms used for training/validation/testing.
###    namely 1. downsampled to 4096Hz, 2. filtered with low cut off equal to 11Hz, 3. tukey window with alpha = 0.8 applied, 4. padded with zero to a pre-set length


### 2. For each test below, one set of magnetorotational waveforms and one set of neutrino -driven waveforms were chosen. 
###    The waveforms of each mechanisms were then duplicated until there are around 400 waveforms for each waveform mechanisms
###    400 pure noise background realisations were then generated, so that the testing data samples were around 1200 in total.



## Result

### 1. These waveforms were taken from Dimmelmeier, et al 2008 (magnetorotational mechanism) and Murphy, et al, 2009 (neutrino-driven mechanism)
### These waveforms are also used in Louge, et al, 2012


### 2. These waveforms were taken from Abdikamalov, et al 2014 (magnetorotational mechanism) and Yakunin, et al, 2015 (neutrino-driven mechanism)

### 3. These waveforms were taken from Scheidegger, et al, 2010 (magnetorotational mechanism) and Andresen, et al, 2016 (neutrino-driven mechanism)
### These waveforms are also used in Rome, et al, 2019






```python
import Image
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import numpy as np
```

## Figure 1. Performance of the CNN on the waveforms from Dimmelmeier, et al 2008 (magnetorotational mechanism) and Murphy, et al, 2009 (neutrino-driven mechanism)

### the left panel shows the results for waveforms from Dimmelmeier, et al 2008
### the right panels shows the results for waveforms from Murphy, et al, 2009
###  the dashed lines show the results for the network of H L V K, the solid lines shows the results for the network of H+ L+ V K

```python
img = mpimg.imread('Dimmelmeier_2008_Murphy_2009_mag.png')
img2 = mpimg.imread('Dimmelmeier_2008_Murphy_2009_neu.png')

fig, ax1 = plt.subplots(1,2, figsize=(41.,41), sharex = False)
gs1 = gridspec.GridSpec(4, 4)
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1], wspace=0.0, hspace=0.0, top=0.95, bottom=0.05, left=0.17, right=0.845) 
ax1[0].imshow(img)
ax1[1].imshow(img2)

for i in range(len(ax1)):
    ax1[i].patch.set_visible(False)
    ax1[i].axis('off')
    #ax1[i].set_aspect('equal')
    
```
